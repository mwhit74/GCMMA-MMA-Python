```python
import numpy as np
from numpy.linalg import norm
from mmapy import mma
```

```python
def obj_fnc(x):
    return x[0][0]**2 + x[1][0]**2 + x[2][0]**2

def grad_of(x):
    return np.transpose(np.array([[2*x[0][0],2*x[1][0],2*x[2][0]]]))

def constraints(x):
    c1 = (x[0][0]-5)**2 + (x[1][0]-2)**2 + (x[2][0]-1)**2 - 9
    c2 = (x[0][0]-3)**2 + (x[1][0]-4)**2 + (x[2][0]-3)**2 - 9
    
    c = np.transpose(np.array([[c1,c2]]))
    
    return c

def grad_const(x):
    gc = np.array([[2*(x[0][0]-5),2*(x[1][0]-2),2*(x[2][0]-1)],
                   [2*(x[0][0]-3),2*(x[1][0]-4),2*(x[2][0]-3)]])
    
    return gc

m = 2
n = 3

eeen = np.ones((n,1))
eeem = np.ones((m,1))
zeron = np.zeros((n,1))
zerom = np.zeros((m,1))

x = np.transpose(np.array([[4,3,2]]))
xmin = 0.001*eeen
xmax = 5*eeen
xold1 = x.copy()
xold2 = x.copy()

low = xmin.copy()
upp = xmax.copy()

a0 = 1
a = zerom.copy()
c = 1000*eeem
d = eeem.copy()
move = 1.0 #default value

kkt_tol = 10e-9
kkt_norm = kkt_tol*10e3

maxiter = 15

f0val = obj_fnc(x)
df0dx = grad_of(x)
fval = constraints(x)
dfdx = grad_const(x)

print(x)
print(x.shape)
print(f0val)
print(f0val.shape)
print(df0dx)
print(df0dx.shape)
print(fval)
print(fval.shape)
print(dfdx)
print(dfdx.shape)

line = f'{1} {x[0]} {x[1]} {x[2]} {f0val} {fval[0]} {fval[1]}'
print(line)

i = 1
j = 1
while kkt_norm > kkt_tol and i < maxiter:
    xmma, ymma, zmma, lam, xsi, eta, mu, zet, s, low, upp = mma.mmasub(m,n,j,x,xmin,xmax,xold1,xold2,f0val,df0dx,fval,dfdx,low,upp,a0,a,c,d,move)
    
    i+=1
    j+=1
    xold2 = xold1.copy()
    xold1 = x.copy()
    x = xmma.copy()
    
    f0val = obj_fnc(x)
    df0dx = grad_of(x)
    fval = constraints(x)
    dfdx = grad_const(x)
    
    residual,kkt_norm,residual_max = mma.kktcheck(m,n,xmma,ymma,zmma,lam,xsi,eta,mu,zet,s,xmin,xmax,df0dx,fval,dfdx,a0,a,c,d)
    
    line = f'{i} {x[0]} {x[1]} {x[2]} {f0val} {fval[0]} {fval[1]}'
    print(line)
```
